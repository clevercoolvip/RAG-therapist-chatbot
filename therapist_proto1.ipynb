# Langchain CSV LOADER

```python
from langchain.document_loaders.csv_loader import CSVLoader

loader = CSVLoader(file_path="therapy_final_data.csv", source_column='Context')
data = loader.load()
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS

embedder = HuggingFaceEmbeddings()
vectordb = FAISS.from_documents(data, embedder)
```

```python
retriever = vectordb.as_retriever()
rdocs = retriever.get_relevant_documents("I am a man so society has high expectations from me but i dont want to feel pressurized, but still i feel pressurized. What to do in this situation?")
rdocs[0]
```
